## 6. Set Up Logging and Error Handling

Configure a reusable logger early so debugging is consistent across scripts and the app.

```python
"""Logging utilities for the project."""

from __future__ import annotations

import logging


def configure_logging(level: int = logging.INFO) -> None:
    """Set up root logging with a readable format."""
    logging.basicConfig(
        level=level,
        format="%(asctime)s | %(levelname)s | %(name)s | %(message)s",
    )


try:
    configure_logging()
except Exception as exc:  # noqa: BLE001
    print(f"Logging setup failed: {exc}")
```

````python
"""Logging utilities for the project."""

from __future__ import annotations

import logging


def configure_logging(level: int = logging.INFO) -> None:
    """Set up root logging with a readable format."""
    logging.basicConfig(
        level=level,
        format="%(asctime)s | %(levelname)s | %(name)s | %(message)s",
    )


try:
    configure_logging()
except Exception as exc:  # noqa: BLE001
    print(f"Logging setup failed: {exc}")
````
